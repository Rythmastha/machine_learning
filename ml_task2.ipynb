# QUESTION2

```python
#question2
#2. Find the % of males scoring above 80 in math and 85 in reading group wise under [parental level of education as bachelor degree
```

```python
male_df=df[(df['gender']=='male') & (df['parental level of education']=="bachelor's degree")]
male_df
```

```python
y=len(male_df[(male_df['math score']>80) & (male_df['reading score'] > 85)])
y
```

```python
x=len(male_df[(male_df['parental level of education']=="bachelor's degree")])
x
```

```python
percentage=100*y/x
percentage
```

# QUESTION 10

```python
#question 10
#10. Using any kind of plot, find whether the data is left skewed or right skewed based on scores of each subject.
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
sns.kdeplot(ax=axes[0],data=df,x='math score')
sns.kdeplot(ax=axes[1],data=df,x='writing score')
sns.kdeplot(ax=axes[2],data=df,x='reading score')
plt.tight_layout()
```

AFTER NOTICING THE GRAPHS,IT APPEARS THAT THE DATA IS RIGHT SKEWED
