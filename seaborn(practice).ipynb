```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scatter plots
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df
```

```python
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
plt.figure(figsize=(15,5))
#hue =color the plot with the column level of education
```

```python
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

```python
#alpha=transparency parameter 
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=1,style='level of education',hue='level of education')
plt.savefig("my_plot.jpg")
```

```python
#distribution plots
#there are three main distribution plots :
    #rug plot
    #histogram
    #KDE plot=kernel density estimation
```

```python
#rug plot=simplest distribution plot and merely adds a dash or tick line for every single value
#y axis doesnt have a meaning for a rug plot
```

```python
#if we count how many ticks there are per various x-ranges ,we can create a histogram
```

```python
plt.figure(figsize=(5,5))
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
#do not use distplot #depricated!
sns.set(style='white')
sns.displot(data=df,x='salary',bins=10,kde=True,rug=True)
```

```python
sns.histplot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='reading score',y='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2')
```

```python
sns.swarmplot(data=df,x='math score',size=3)
```

```python
#comparison plots
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex',alpha=1)
```

```python
sns.pairplot(data=df)
```

```python
#seaborn grid plots
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch')
```

```python
g=sns.PairGrid(df)
g=g.map_upper(sns.scatterplot)
g=g.map_lower(sns.kdeplot)
```
