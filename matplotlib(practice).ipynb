```python
a=np.linspace(0,10,11)
a
```

```python
b=a**4
b
```

```python
x=np.arange(0,10)
y=2*x
x
```

```python
y
```

```python
fig=plt.figure()
#large axes
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('power of 4')

#small axes
axes2=fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x,y)
```

```python
type(fig)
```

```python
fig=plt.figure(dpi=200)#for clarity of graph
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(figsize=(10,3))#10 by 3 inches size of the graph
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
#sublpots functionality
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.show()
```

```python
axes
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][1].plot(x,y)
axes[1][1].plot(a,b)
plt.show()
```

```python
#can also apply loop
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()#ensures no overlapping of axes
```

```python
#styling
#legend
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs x')
ax.plot(x,x**2,label='square')
ax.legend()
plt.show()
```

```python
#colors and styles
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple',linewidth=2)
plt.show()
```
